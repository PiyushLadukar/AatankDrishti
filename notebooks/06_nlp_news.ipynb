# Step 6: NLP News Analysis

Compare media coverage distribution vs actual GTD data.
Does news over/under-represent certain ideologies?

```python
import sys
sys.path.append('../src')
import pandas as pd
from nlp.news_analyzer import NewsAnalyzer

gtd_df = pd.read_csv('../data/processed/gtd_processed.csv', low_memory=False)
print('GTD loaded')
```

```python
# Option A: Use synthetic articles (no API key needed)
analyzer = NewsAnalyzer(use_llm=False)
articles = analyzer.generate_sample_articles(n=500)
print(f'Sample articles: {len(articles)}')
articles.head(3)
```

```python
# Option B: Fetch real news (requires News API key in .env)
# import os
# from newsapi import NewsApiClient
# newsapi = NewsApiClient(api_key=os.getenv('NEWS_API_KEY'))
# top = newsapi.get_everything(q='terrorism', language='en', page_size=100)
# articles = pd.DataFrame([{'text': a['title'] + ' ' + (a['description'] or '')} for a in top['articles']])
```

```python
# Analyze articles
results = analyzer.analyze(articles)
print('\nPredicted ideology distribution in news:')
print(results['predicted_ideology'].value_counts().to_string())
```

```python
# Compare media vs GTD
analyzer.compare_to_gtd(results, gtd_df)

# Evaluate accuracy on sample (we know true_ideology for synthetic articles)
if 'true_ideology' in results.columns:
    from sklearn.metrics import accuracy_score
    acc = accuracy_score(results['true_ideology'], results['predicted_ideology'])
    print(f'\nKeyword classifier accuracy on sample articles: {acc:.2%}')
```

```python
results.to_csv('../data/processed/news_analysis.csv', index=False)
print('Saved → data/processed/news_analysis.csv')
print('\nProject complete! Run backend/app.py + frontend to view the dashboard.')
```
